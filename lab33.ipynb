```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.stats
```

```python
def U(x, i, j, n):
  sum1, sum2 = 0, 0

  if 0 <= i-1:
    sum1 += x[i-1][j]
  if j+1 < n:
    sum2 += x[i][j+1] 
  if i+1 < n:
    sum1 += x[i+1][j]
  if 0 <= j-1:
    sum2 += x[i][j-1]
  return math.exp( - x[i][j] * beta2 * sum2 - x[i][j] * beta1 * sum1 )
```

```python
n = 10**2
m = 10**5

x = 2*np.random.randint(2, size=(n,n)) - 1

step = 0
beta1 = 1
beta2 = 1.5
t = 0
```

```python
while step <= m: 
  
  u = random.uniform( 0, 1 )
  i = random.randint( 0, n-1 )
  j = random.randint( 0, n-1 )
  t = -x[i][j]

  alpha = U(x, i, j, n)
  if u < alpha:
    x[i][j] = t

  if step == 0 or step == m:
    fig, ax = plt.subplots()
    ax.imshow(x, cmap='binary')
    plt.show()

  step += 1
```
